In this notebook I will be using image segmentation to identify where the phone is. Because the phone is in different angles I will be segmenting the phone using a rectangle about the center. 

```python
import os
from os import walk, path, listdir
from os.path import isfile, join
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import math
from PIL import Image

#get current directory 
current_dir = os.getcwd()

#find phone directory
find_phone_dir = current_dir + '\\find_phone'

#get the file names in the find phone directory
filenames = [f for f in listdir(find_phone_dir) if isfile(join(find_phone_dir, f))]

#get the image names
labels_file = filenames.pop(len(filenames)-1)
num_samples = len(filenames)

#print lists
print(labels_file, '\n')
print(filenames)
print('\nNumber of files: ', num_samples)

#get the label files
label = open(find_phone_dir + '\\' + str(labels_file), 'r')
labels = label.read()
label_list = labels.split('\n')

#delete the last element ('')
label_list.pop()
print('\nThe number of coordinates and labels availabel: ', len(label_list), '\n')

print(label_list)

#add file name and coordinates to a dictionary 
coords = {}

for name in label_list:
    
    name_split = name.split(' ')
    coords[name_split[0]] = tuple((float(name_split[1]), float(name_split[2])))
```

```python
#resize images to 224 , 224 , 3
from PIL import Image

#create a new Xmatrix
X = np.zeros([len(filenames), 320, 480, 3])
Y = np.zeros([len(filenames), 2])

#mask
X_seg = np.zeros([len(filenames), 320, 480, 1])

for i in range(len(filenames)):
    
    im = Image.open(find_phone_dir + '\\' + filenames[i])
    image = im.resize((480, 320))
    
    pixels = np.asarray(image)
    
    #convert the coordinates to new resized coordinates
    Y[i, 0] = coords[filenames[i]][0]
    Y[i, 1] = coords[filenames[i]][1]
    
    #append to X
    X[i, :] = pixels
    
    #get x direction phone mask
    x_low = math.floor((coords[filenames[i]][0] - 0.05) * 480)
    x_high = math.floor((coords[filenames[i]][0] + 0.05)* 480)
    
    #get y direction phone mask
    y_low = math.floor((coords[filenames[i]][1] - 0.05) * 320)
    y_high = math.floor((coords[filenames[i]][1] + 0.05) * 320)    
    
    #masked images
    X_seg[i, y_low:y_high, x_low:x_high, 0] = 1
```

```python
#show a masked and original images side by side
rand_pic = np.random.choice(np.arange(0, 129))

fig, ax = plt.subplots(1, 2, figsize = (12, 4))
ax[0].imshow(X[rand_pic, :, :, :].astype('uint8'))
ax[0].set_title('Original Image')
ax[1].imshow(X_seg[rand_pic, :, :, :].astype('uint8'))
ax[1].set_title('Masked Image')
```

Now create the unet model and predict the location of the phone. 




```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Dropout 
from tensorflow.keras.layers import Conv2DTranspose
from tensorflow.keras.layers import concatenate
```

```python
#split the data
#split to train and test sets
X_train = X[:110]/255
X_train = X_train.astype('float32')
Y_train = X_seg[:110].astype('uint8')
X_test = X[110:len(X)]/255
X_test = X_test.astype('float32')
Y_test = X_seg[110:len(X)].astype('uint8')
```

```python
#downsampling block
def conv_block(inputs = None, n_filters = 32, max_pooling = True):
    
    #conv1
    conv = Conv2D(n_filters, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(inputs)
    #conv2
    conv = Conv2D(n_filters, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv)
    
    if max_pooling:
        
        next_layer = MaxPooling2D((2, 2))(conv)
        
    else:
        
        next_layer = conv
        
    skip = conv
    
    return next_layer, skip

#upsampling part
def upsampling_block(expansive_input, contractive_input, n_filters = 32):
    
    #upsampling the input
    up = Conv2DTranspose(n_filters, (3, 3), strides = (2, 2), padding = 'same')(expansive_input)
    
    #merge them together
    merge = concatenate([up, contractive_input], axis = 3)
    
    conv = Conv2D(n_filters, (3, 3), activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(merge)
    
    conv = Conv2D(n_filters, (3, 3), activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv)
    
    return conv
```

```python
#build the unet model
def unet_model(input_size = (320, 480, 3), n_filters = 32, n_classes = 2):
    
    inputs = Input(input_size)
    
    #convolution lock 1
    cblock1 = conv_block(inputs)
    cblock2 = conv_block(cblock1[0], n_filters * 2)
    cblock3 = conv_block(cblock2[0], n_filters * 4)
    cblock4 = conv_block(cblock3[0], n_filters * 8)
    cblock5 = conv_block(cblock4[0], n_filters * 16, max_pooling = False)
    
    #upsmapling
    ublock6 = upsampling_block(cblock5[0], cblock4[1], n_filters * 8)
    ublock7 = upsampling_block(ublock6, cblock3[1], n_filters * 4)
    ublock8 = upsampling_block(ublock7, cblock2[1], n_filters * 2)
    ublock9 = upsampling_block(ublock8, cblock1[1], n_filters)
    
    conv9 = Conv2D(n_filters, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(ublock9)
    
    conv10 = Conv2D(n_classes, (1, 1), padding = 'same')(conv9)
    
    model = tf.keras.Model(inputs = inputs, outputs = conv10)
    
    return model
```

```python
#initialize the model
unet = unet_model()
unet.summary()
```

```python
#compile


unet.compile(optimizer = 'adam', loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True), metrics = ['accuracy'])
```

```python
#early stopping
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience= 4)

#fit the model
model_hist = unet.fit(x = X_train, y = Y_train, epochs = 8, callbacks = callback)
```

```python
#predict the mask using the model
predictions = unet.predict(X_test)
```

```python
#plot original pic, true mask and predicted mask

for i in range(len(X_test)):
    
    #predicted im 
    im = predictions[i, :, :, :]
    im = np.argmax(im, axis = 2)
    
    #plot the three images
    fig, ax = plt.subplots(1, 3, figsize = (12, 4))
    ax[0].imshow(X[110 + i, :, :, :].astype('uint8'))
    ax[0].set_title('Original Image')
    ax[1].imshow(Y_test[i, :, :, :].astype('uint8'), cmap = 'gray')
    ax[1].set_title('True Mask Created')
    ax[2].imshow(im.astype('uint8'), cmap = 'gray')
    ax[2].set_title('Predicted Mask by the Model')
```

The predicted mask does not look exactly like the true mask, but it is segementing the phone slightly. Given that the mask was created by making a rectangle about the coordinates of the phone, Unet algorithm seems to segment the phone pretty well. 

In the future if we segment the images better, the model will be able to segment the phone pretty well, given the results of the current version.
